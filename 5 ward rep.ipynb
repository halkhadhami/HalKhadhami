### MapReduce python code

```python
import string
```

```python
file = open("db/Leonardo.txt")
data = file.read()
```

```python
data = data.translate(str.maketrans('', '', string.punctuation))
data = data.lower()
data = ''.join([i for i in data if not i.isdigit()])
data = data.strip()
words = data.split()
all_w_count = len(words)
print('Total words in the text file =',all_w_count)
```

```python
def mapp(words):
    lis_of_5w_lists_map = dict()
    _5w_long = all_w_count - 4
    for i in range(_5w_long):
        start_indx = i
        _5w_list = tuple(words[start_indx : start_indx + 5])
        key = _5w_list
        value = 1
        
        lis_of_5w_lists_map.update({key : value})
    
    #print(lis_of_5w_lists)
    return lis_of_5w_lists_map
```

```python
def group(lis_of_5w_lists):
    lis_of_5w_group = dict()

    for item in lis_of_5w_lists:
        key = item
        value = 1
        if key in lis_of_5w_group:
            lis_of_5w_group[key].append(value)
        
        else:
            lis_of_5w_group.update( {key : value} )
    #print(lis_of_5w_group)
    return lis_of_5w_group  
```

```python
#print(group(mapp(words)))
```

```python
def reduce(lis_of_5w_lists):
    lis_of_5w_reduce = dict()

    for item in lis_of_5w_lists:
        key = item
        value = 1
        if key in lis_of_5w_reduce:
            lis_of_5w_reduce[key] += value
        
        else:
            lis_of_5w_reduce.update( {key : value} )
    #print(lis_of_5w_group)
    return lis_of_5w_reduce  
```

```python
print(reduce(group(mapp(words))))
```
